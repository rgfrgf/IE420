```python
# module imports
import math
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import time
from scipy.stats import norm
```

```python
def nested():
    return defaultdict(int)
def UDCal(sigma,T,N):
    step=T/N
    u=math.exp(sigma*math.sqrt(step))
    d=math.exp(-1*sigma*math.sqrt(step))
    #print('u=',u,'d=',d)
    return u,d
def finalCal(Option,K,S0,u,d,N):
    finalnode=defaultdict(int)
    for i in range(0,N+1):#i is the count of going up, range 0~N, N times max, 0 times min
        if Option=='C':
            finalnode[i]=max(S0*(u**i)*(d**(N-i))-K,0)
        if Option=='P':
            finalnode[i]=max(K-S0*(u**i)*(d**(N-i)),0)
    return finalnode
def riskneutral(r,q,u,d,T,N):
    step=T/N
    p=(math.exp((r-q)*step)-d)/(u-d)
    return p
def CRR(Option,K,T,S0,sigma,r,q,N,Exercise='E'):#T in years #calculation, price calculation, early exercise boundary not mentioned
    start=time.time()
    u,d=UDCal(sigma,T,N)
    finalnode=finalCal(Option,K,S0,u,d,N) #finalnode is a defaultdict with the key as times going up, value as the value at that situation
    #print('finalnode=',finalnode)
    step=T/N
    discount=math.exp(-1*step*r)
    p=riskneutral(r,q,u,d,T,N)
    #print('riskneutral=',p)
    binomialtree=defaultdict(nested)
    binomialtree[N]=finalnode
    #print('entering if sentence')
    if Exercise=='E'or Exercise=='e':
        for each_step in range(N-1,-1,-1):
            for risetime in range(0,each_step+1): #e.g. step=N-1, maximum risetime=N-1
                binomialtree[each_step][risetime]=discount*((1-p)*binomialtree[each_step+1][risetime]+p*binomialtree[each_step+1][risetime+1])
        end=time.time()
        return binomialtree,end-start
    if Exercise=='A'or Exercise=='a':
        for each_step in range(N-1,-1,-1):
            for risetime in range(0,each_step+1): #e.g. step=N-1, maximum risetime=N-1
                continual_value=discount*((1-p)*binomialtree[each_step+1][risetime]+p*binomialtree[each_step+1][risetime+1])
                exercise_value=S0*(u**risetime)*(d**(each_step-risetime))-K if (Option=='C' or Option=='c') else K-S0*(u**risetime)*(d**(each_step-risetime))
                binomialtree[each_step][risetime]=max(exercise_value,continual_value)                   
        end=time.time()
        return binomialtree,end-start
                        
        
```

```python
def BSmodel(Option,K,T,S0,sigma,r,q):
    value=0
    start=time.time()
    d1=(math.log(S0/K)+(r-q+0.5*sigma**2)*T)/(sigma*math.sqrt(T))
    d2=d1-sigma*math.sqrt(T)
    if Option=='C' or Option=='c':
        value=S0*math.exp(-1*q*T)*norm.cdf(d1)-K*math.exp(-1*r*T)*norm.cdf(d2)
    if Option=='P' or Option=='p':
        value=-1*S0*math.exp(-1*q*T)*norm.cdf(-1*d1)+K*math.exp(-1*r*T)*norm.cdf(-1*d2)
    end=time.time()
    return value,end-start
```

```python
#Question 2 verifying that the binomial model converges to B-S model with higher step
def Q2(N):
    K=100
    T=1
    S0=100
    sigma=0.2
    r=0.05 
    q=0.04
    BS_Result,BS_Time=BSmodel('C',K,T,S0,sigma,r,q)
    print(BS_Result,BS_Time)
    Binomial_Result=[]
    X=[]
    Bino_time=[]
    for i in range(1,N+1):
        Bino_Tree,Bino_Time=CRR('C',K,T,S0,sigma,r,q,i,'E')
        Bino_Result=Bino_Tree[0][0]
        Binomial_Result.append(Bino_Result)
        Bino_time.append(Bino_Time)
        #print('Bino_Result',Bino_Result,'Bino_Time',Bino_Time)
        X.append(i)
    Bino_time=np.array(Bino_time)
    print('average time',np.mean(Bino_time),'average price',np.mean(Bino_Result))
    Binomial_Result=np.array(Binomial_Result)
    BS=np.full((N,),BS_Result)
    
    plt.figure(figsize=(20,10))
    plot1=plt.plot(X,Binomial_Result,'r',label='BinomialTree')
    plot2=plt.plot(X,BS,'g',label='B-S Result')
    plt.legend(loc="upper left")
    plt.xlabel('Steps')
    plt.ylabel('Option Price')
    plt.show()
Q2(200)
```

```python
def findingN():
    K=100
    sigma=0.2
    r=0.05
    q=0
    S0=100
    Maturity=[]
    Optionprice=defaultdict(nested)
    for i in range(1,12+1):
        Maturity.append(i/12)
    for i in Maturity:
        print('matruity=',i)
        for N in range(1,5000,100):
            NewValue,run_time=CRR('P',K,i,S0,sigma,r,q,N,'A')
            Optionprice[i*12][N]=NewValue[0][0]#first key=month i, second key[step]
    return Optionprice
```

```python
def plotwithS0(Option,K,T,sigma,r,q,N):
    Bino_price=[]
    Spot_price=[]
    time_loop=[]
    for S0 in range(1,120):
        start=time.time()
        Bino_Tree,Bino_Time=CRR(Option,K,T,S0,sigma,r,q,N,'A')
        Bino_price.append(Bino_Tree[0][0])
        Spot_price.append(S0)
        end=time.time()
        time_loop.append(end-start)
    Spot_price=np.array(Spot_price)
    Bino_price=np.array(Bino_price)
    plt.figure(figsize=(20,10))
    plot1=plt.plot(Spot_price,Bino_price,'r',label='BinomialTree')
    plot2=plt.plot(Spot_price,K-Spot_price,'b')
    plt.legend(loc="upper right")
    plt.xlabel('SpotPrice')
    plt.ylabel('Option Price')
    plt.show()
    return time_loop
#input paramter for Q3 first part
K=100
T=1
sigma=0.2
r=0.05
q=0
N=1201
start=time.time()
plotwithS0('P',K,T,sigma,r,q,N)
end=time.time()
print('time to plot the S0',end-start)
```

```python
def EarlyExercise(Option,K,T,sigma,r,q,N):
    start=time.time()
    Exercise='A'
    difference1=[]#first loop with bigger step
    difference2=[]#second loop with smaller step
    Time_used1=[]
    Time_used2=[]
    Initial_price1=[]
    Initial_price2=[]
    if Option=='P'or Option=='p':
        for S0 in range(50,K,1):
            bino_tree,cal_time=CRR(Option,K,T,S0,sigma,r,q,N,Exercise)
            Time_used1.append(cal_time)
            option_price=bino_tree[0][0]
            intrinsic_value=max(0,K-S0) 
            if (option_price-intrinsic_value>0.005):
                break
            difference1.append(option_price-intrinsic_value)
            #print('at input S0=',S0,'option_price=',option_price,'intrinsic_value=',intrinsic_value,option_price-intrinsic_value)
            Initial_price1.append(S0) 
        difference1=np.array(difference1)
        Time_used1=np.array(Time_used1)
        Initial_price1=np.array(Initial_price1)
        lower_value=max(Initial_price1[np.all([difference1<0.005],axis=0)]) 
        #print('diffrence in 1st loop',difference1,'corresponding maximum',lower_value)
        for S0 in np.arange(lower_value,lower_value+1,0.01):
            bino_tree,cal_time=CRR(Option,K,T,S0,sigma,r,q,N,Exercise) 
            Time_used2.append(cal_time)
            option_price=bino_tree[0][0]
            intrinsic_value=max(0,K-S0)
            if (option_price-intrinsic_value>0.005):
                break
            difference2.append(option_price-intrinsic_value)
            Initial_price2.append(S0)
            #print('at input S0=',S0,'option_price=',option_price,'difference=',option_price-intrinsic_value)          
        difference2=np.array(difference2)
        Time_used2=np.array(Time_used2)
        Initial_price2=np.array(Initial_price2)
        lower_value2=max(Initial_price2[np.all([difference2<0.005],axis=0)]) 
        #print('difference in 2nd loop',difference2,'corresponding maximum',lower_value2)
        end=time.time()
        print('time used to run 2loops',end-start)
    return difference1,Time_used1,Initial_price1,difference2,Time_used2,lower_value2
```

```python
def plotprice(PRICE):
    Maturity=[]
    for T in range(1,13):
        Maturity.append(T)
    Maturity=np.array(Maturity)
    plt.plot(Maturity,PRICE,'r',label='critical spot price')
    plt.legend(loc="upper right")
    plt.xlabel('Maturity(Month)')
    plt.ylabel('critical stock price S* ')
    plt.show()
```

```python
K=100
sigma=0.2
r=0.05
q=0.0
N=1201
PRICE_q0=np.array([])
average_time=np.array([])
Time_loop=np.array([])
for T in range(1,13):
    T=T/12
    difference1,Time_used1,Initial_price1,difference2,Time_used2,lower_value2=EarlyExercise('P',K,T,sigma,r,q,N)
    print('at maturity',T*12,'month',lower_value2)
    PRICE_q0.append(lower_value2)
    Time_loop=np.append(Time_loop,Time_used1)
    Time_loop=np.append(Time_loop,Time_used2)
    average_time.append(np.mean(Time_loop))
```
